```python
import pandas as pd
git_url = 'https://raw.githubusercontent.com/nafis-git/Data_analysis_python/main/Python_Fundementals/data/happiness.csv'
df = pd.read_csv(git_url)
```

```python
df
```

```python
df_av_country = df.groupby(['Country name'])['Healthy life expectancy at birth','Log GDP per capita'].mean()
```

```python
df_av_country
```

```python
git_url = 'https://raw.githubusercontent.com/nafis-git/Data_analysis_python/main/Python_Fundementals/data/world_population.csv'
world_pop =pd.read_csv(git_url, skiprows=[0,1,2], header = 1)
world_pop.head()
```

```python
# world_pop.info()
```

```python
world_pop.drop(['Country Code','Indicator Name','Indicator Code'], axis = 1, inplace = True)
world_pop.head()
```

```python
world_pop_08_18 = world_pop.loc[:,'2008':'2018']
```

```python
world_pop_08_18.insert(0, 'country',  world_pop['Country Name'])
world_pop_08_18.insert(1, 'av_pop',  world_pop_08_18.mean(axis = 1))
```

```python
world_pop_08_18
```

```python
dg = world_pop_08_18.iloc[:,:2]
dg.set_index('country', inplace = True)
dg.sort_values(by='country', inplace = True)
```

```python
dg
```

```python
# lets join two dataframe of world happiness and world population (df_av_country,dg )
df_joint = df_av_country.join(dg, on=None, how='right').dropna()
df_joint.columns = ['life_exp', 'log_gdp', 'ave_pop']
df_joint
```

```python
gdp_cap = df_joint['log_gdp']
life_exp = df_joint['life_exp']
pop = map(lambda x:int(x), np.array(df_joint['ave_pop'])/1000000)
```

```python
pop_array = np.array(list(pop))
pop_array.shape
```

```python
# we want to color countries based on their continents so we add a table indicating contienet of countries
```

```python
urls= 'https://raw.githubusercontent.com/dbouquin/IS_608/master/NanosatDB_munging/Countries-Continents.csv'
df_contines = pd.read_csv(urls, index_col = 'Country')
```

```python
df_contines.sort_values(by = 'Country', inplace = True)
df_contines.head()
```

```python
df_contines['Continent'].unique()
```

```python
df_contries_continensts = df_joint.join(df_contines, on = None, how= 'left')
```

```python
df_contries_continensts.isnull().sum().sum()
```

```python
df_contries_continensts.loc[df_contries_continensts['Continent'].isnull()].index
```

```python
df_contries_continensts.loc[df_contries_continensts['Continent'].isnull()].squeeze()
```

```python
index_Burkina = np.where(df_contries_continensts.index == 'Burkina Faso')[0][0]
index_Kosovo = np.where(df_contries_continensts.index == 'Kosovo')[0][0]
index_Myanmar = np.where(df_contries_continensts.index == 'Myanmar')[0][0]
index_States = np.where(df_contries_continensts.index == 'United States')[0][0]
```

```python
index_Burkina
```

```python
# to fill the nan values with correct values
df_contries_continensts.iloc[index_Burkina,-1] = 'Africa'
df_contries_continensts.iloc[index_Kosovo,-1] = 'Europe'
df_contries_continensts.iloc[index_Myanmar,-1] = 'Asia'
df_contries_continensts.iloc[index_States,-1] = 'North America'
```

```python
df_contries_continensts['Continent'] = df_contries_continensts['Continent'].astype('category')
```

```python
df_contries_continensts.info()
```

```python
dict_c = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'North America':'yellow',
    'South America':'orange',
    'Oceania':'black'
}
```

```python
# to make a list based on the continent
colors = [*map(dict_c.get, df_contries_continensts['Continent'])]
```
