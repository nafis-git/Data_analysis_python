## emire state case study
You throw a die one hundred times.

- If it's 1 or 2 you'll go one step down.
- If it's 3, 4, or 5, you'll go one step up.
- If you throw a 6, you'll throw the die again and will walk up the resulting number of steps.
- you can not go lower than step number 0. And also, you admit that you're a bit clumsy and have a chance of 0.1% of falling down the stairs when you make a move. Falling down means that you have to start again from step 0. With all of this in mind, you bet with your friend that you'll reach 60 steps


```python
np.random.seed(123)
# Starting step
steps = [0]
# Roll the dice for 00 times
for nth in range(100):
    dice = np.random.randint(1,7)
# Finish the control construct
    if (dice <= 2) & (steps[-1] != 0):
        steps.append(steps[-1]-1)
    elif dice in [3,4,5]:
        steps.append(steps[-1]+1)
    elif dice == 6 :
        steps.append(steps[-1] + np.random.randint(1,7))

# Print out dice and step
print(f'last step is {steps[-1]} and dice is {dice}')
```

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Plot random_walk

plt.plot(steps)
plt.xlabel('step')
plt.ylabel('accumelated_step')
# Show the plot
plt.show()
```

```python
# Simulate random walk 10 times
np.random.seed(123)
# Starting step
all_walks= []
final_steps = []
for i in range(250) :
    steps = [0]
    # Roll the dice for 00 times
    for nth in range(100):
        step = steps[-1]
        dice = np.random.randint(1,7)
    # Finish the control construct
        if (dice <= 2) & (step != 0):
            steps.append(step-1)
        elif dice in [3,4,5]:
            steps.append(step+1)
        elif dice == 6 :
            steps.append(step + np.random.randint(1,7))
        # Implement clumsiness
        if np.random.rand()<.001 :
            step = 0
    final_steps.append(steps[-1])
    all_walks.append(steps)

# print(final_steps)        
```

```python
# plot all walks 
for walk in all_walks:
    plt.plot(np.array(walk).T)
```

### if we simulate for 500 times, To calculate the chance that this end point is greater than or equal to 60, you can count the number of integers in ends that are greater than or equal to 60 and divide that number by 500, the total number of simulations.

```python
# Simulate random walk 10 times
np.random.seed(123)
# Starting step
all_walks= []
final_steps = []
for i in range(500) :
    steps = [0]
    # Roll the dice for 00 times
    for nth in range(100):
        step = steps[-1]
        dice = np.random.randint(1,7)
    # Finish the control construct
        if (dice <= 2) & (step != 0):
            steps.append(step-1)
        elif dice in [3,4,5]:
            steps.append(step+1)
        elif dice == 6 :
            steps.append(step + np.random.randint(1,7))
        # Implement clumsiness
        if np.random.rand()<.001 :
            step = 0
    final_steps.append(steps[-1])
    all_walks.append(steps)
```

```python
sum(np.array(final_steps)>=60)
```

```python
sum(np.array(final_steps)>=60)/500
```
