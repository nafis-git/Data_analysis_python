```python
# import required libraries
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import datetime as dt
from sklearn import preprocessing
```

```python
data = yf.download('LUV TSLA', start = '2020-01-01',end = dt.date.today())
```

```python
data
```

```python
data_arry= data.values
data_arry
```

```python
#Normalize data with MinMaxScaler using SK-Learn
min_max_scaled = preprocessing.MinMaxScaler()
scaled_data= min_max_scaled.fit_transform(data_arry)
df_scaled_data = pd.DataFrame(scaled_data)
df_scaled_data.columns = data.columns
df_scaled_data.index = data.index
df_scaled_data.head()
```

```python
#Normalize data with (data-data.mean())/data.std()
normalized_data = (data - data.mean())/data.std()
normalized_data.head()
```
